# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

# Read Dataset

```python
data = pd.read_csv("Netflix Data.csv")
```

```python
data.head(10)
```

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.set_index('Date')
data.head()
```

# First we shall analyze the data by plotting the data over the whole duration
and then over different periods to look for patterns.
sns.lineplot(x= data.index, y=data['Volume'], label ='Volume')
plt.title('Volume of stock versus time')

```python
data.plot(y= ['High','Close','Open'], title = 'Netflix stock price')
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3,figsize=(15,10))
data.groupby(data.index.day).mean().plot(y='Volume',ax=ax1,xlabel='Day')
data.groupby(data.index.day).mean().plot(y='Volume',ax=ax2,xlabel='Month')
data.groupby(data.index.day).mean().plot(y='Volume',ax=ax3,xlabel='Year')
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3,figsize=(15,10))
data.groupby(data.index.day).mean().plot(y=['High','Close','Low'],ax=ax1,xlabel='Day')
data.groupby(data.index.month).mean().plot(y=['High','Close','Low'],ax=ax2, xlabel='Month')
data.groupby(data.index.year).mean().plot(y=['High','Close','Low'], ax=ax3, xlabel='Year')
```

# Dates with Highest Stock price

```python
data = data.sort_values(by = 'High', ascending = False).head(10)
data['High']
```

# Dates with Lowest Stock Price

```python
data = data.sort_values(by = 'Low', ascending = True).head(10)
data['Low']
```

```python
fig,axes = plt.subplots(nrows= 1,ncols=2,sharex = True, figsize=(12,5))
fig.suptitle('High & Low Values stock per period of time ', fontsize =18)
sns.lineplot(ax = axes[0], y= data['High'], x=data.index, color='green')
sns.lineplot(ax = axes[1], y= data['Low'], x=data.index, color='red')
```
